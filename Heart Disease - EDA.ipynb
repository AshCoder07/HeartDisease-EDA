## Distribution Charts for Numerical data

```python
sns.displot(df["Age"],kde=True)
```

```python
sns.distplot(df["Age"],hist=True,kde=False)
```

```python
sns.distplot(df["Age"],rug=True,kde=False,hist=False)
```

```python
sns.distplot(df["Age"],hist=True,kde=True,rug=True)
```

```python
sns.distplot(df["Age"],bins=10)
```

```python
sns.displot(df["RestingBP"],kde=True,color="red")
```

```python
sns.distplot(df["Cholesterol"],kde=True,color="red")
```

```python
sns.distplot(df["FastingBS"],kde=True,color="red")
```

```python
sns.distplot(df["MaxHR"],kde=True,color="red")
```
